# Relembrando funções com retorno e parâmetro e que usem laços de repetição simples


## Essa função retorna a soma de números impares

```python
def soma_impares(n):
    soma = 0
    for i in range(1, n + 1):
        if i % 2 != 0:   # testa se é ímpar
          soma += i
    return soma
```

```python
# Exemplo de uso
resultado = soma_impares(10)
print("A soma dos ímpares é:", resultado)
```

## Relebrando mais uma função  que usa laço de reptição e dicionários

```python
def cadastrar_produtos(qtd):
    produtos = {}

    for i in range(qtd):
        print(f"\nProduto {i+1}:")
        nome = input("Digite o nome do produto: ")
        preco = float(input("Digite o preço: "))

        produtos[nome] = preco  # armazena no dicionário

    return produtos

# Exemplo de uso
lista_produtos = cadastrar_produtos(3)
print("\nProdutos cadastrados:")
print(lista_produtos)
```

# relembrando agora uma função que usa condicionais

```python
def verificar_temperatura(t):
    if t < 0:
        return "Muito frio"
    elif t <= 25:
        return "Temperatura agradável"
    else:
        return "Está quente"
```

1) Crie uma função em Python, com parâmetro e com retorno, que receba um número inteiro n e calcule o produto acumulado dos números de 1 até n.

A função deve funcionar assim:

* Ela começa com o valor inicial igual a 1 e o valor final pelo parâmetro digitador na função;

* Em seguida, multiplica esse valor por todos os números de 1 até n, acumulando o resultado a cada passo.

Exemplo do funcionamento interno quando o usuário informa o número 5:

5 X 1 = 5

5 X 2 =10

10 X 3 = 30

30 X 4 = 120

funcaomultiplica(5)

Sabemos que ele não irá até 5 pois dentro de um laço de repetição o ultimo valor dentro de um range não é considerado.


```python
def multiplicacao_sequencial(num):
    
    resultado = num
    print(f"Iniciando com: {resultado}")
    
    for i in range(2, num + 1):
        resultado *= i  # Multiplica o resultado atual pelo número da iteração
        print(f"{num} x {i} = {resultado}")

    return resultado


```

```python
multiplicacao_sequencial(7)
```
