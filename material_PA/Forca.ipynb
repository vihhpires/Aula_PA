# Atividade - Jogo da forca

**Descrição:**
Você vai criar um programa que:

Escolhe aleatoriamente uma palavra secreta de uma lista.

Permite que o jogador tente adivinhar as letras.

Mostra o progresso da palavra e as tentativas erradas.

Encerra o jogo quando o jogador vencer ou errar 6 vezes.



**Regras:**
O programa deve ser dividido em múltiplas funções:

escolher_palavra(): retorna uma palavra aleatória.

mostrar_palavra(palavra, letras_certas): mostra a palavra com _ nos lugares ainda não descobertos.

tentar_letra(letra, palavra, letras_certas, letras_erradas): atualiza as listas de acertos e erros.

verificar_vitoria(palavra, letras_certas): retorna True se o jogador ganhou.

jogo_da_forca(): função principal que controla o jogo.

```python
import random

# Lista de palavras para o jogo
palavras = ["tv", "linda", "carambola", "inteligencia", "iphone", "computador"]

def escolher_palavra():
    """Escolhe uma palavra aleatória da lista."""
    return random.choice(palavras)

def mostrar_palavra(palavra, letras_certas):
    """Mostra a palavra com _ nos lugares ainda não descobertos."""
    return ''.join([letra if letra in letras_certas else '_' for letra in palavra])

def tentar_letra(letra, palavra, letras_certas, letras_erradas):
    """Tenta a letra fornecida e atualiza as listas de acertos e erros."""
    if letra in palavra:
        letras_certas.append(letra)
        return True
    else:
        letras_erradas.append(letra)
        return False

def verificar_vitoria(palavra, letras_certas):
    """Verifica se o jogador venceu."""
    return all(letra in letras_certas for letra in palavra)

def jogo_da_forca():
    """Função principal que controla o jogo da forca."""
    palavra = escolher_palavra()
    letras_certas = []
    letras_erradas = []
    tentativas = 0
    max_tentativas = 6

    print("Bem-vindo ao jogo da Forca!")

    while tentativas < max_tentativas:
        print("\nPalavra: ", mostrar_palavra(palavra, letras_certas))
        print(f"Tentativas erradas: {', '.join(letras_erradas)}")
        print(f"Você tem {max_tentativas - tentativas} tentativas restantes.")

        letra = input("Tente uma letra: ").lower()

        # Verifica se a entrada é uma letra única
        if len(letra) != 1 or not letra.isalpha():
            print("Por favor, insira apenas uma letra válida.")
            continue

        if letra in letras_certas or letra in letras_erradas:
            print("Você já tentou essa letra antes.")
            continue

        # Verifica se a letra está correta ou não
        if not tentar_letra(letra, palavra, letras_certas, letras_erradas):
            tentativas += 1

        # Verifica se o jogador venceu
        if verificar_vitoria(palavra, letras_certas):
            print("\nParabéns pelo esforço, você ganhou!!!")
            print("A palavra era:", palavra)
            break
    else:
        print("\nVocê perdeu! A palavra era:", palavra)

# Chama a função principal para iniciar o jogo
jogo_da_forca()
```
